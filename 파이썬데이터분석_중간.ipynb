```python
# ğŸ“š ë¼ì´ë¸ŒëŸ¬ë¦¬ ë¶ˆëŸ¬ì˜¤ê¸°
import pandas as pd

# ğŸ“‚ ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
travel = pd.read_csv('/content/drive/MyDrive/tn_travel_ì—¬í–‰_H.csv', encoding='utf-8-sig')
transport = pd.read_csv('/content/drive/MyDrive/tn_mvmn_consume_his_ì´ë™ìˆ˜ë‹¨ì†Œë¹„ë‚´ì—­_H.csv', encoding='utf-8-sig')
lodging = pd.read_csv('/content/drive/MyDrive/tn_lodge_consume_his_ìˆ™ë°•ì†Œë¹„ë‚´ì—­_H.csv', encoding='utf-8-sig')
activity_consume = pd.read_csv('/content/drive/MyDrive/tn_activity_consume_his_í™œë™ì†Œë¹„ë‚´ì—­_H.csv', encoding='utf-8-sig')
activity = pd.read_csv('/content/drive/MyDrive/tn_activity_his_í™œë™ë‚´ì—­_H.csv', encoding='utf-8-sig')
visit = pd.read_csv('/content/drive/MyDrive/tn_visit_area_info_ë°©ë¬¸ì§€ì •ë³´_H.csv', encoding='utf-8-sig')
codeB = pd.read_csv('/content/drive/MyDrive/tc_codeb_ì½”ë“œB.csv', encoding='utf-8-sig')

# ğŸ“‘ í•„ìš”í•œ ì—´ ì¶”ì¶œ
travel = travel[['TRAVEL_ID', 'TRAVEL_PURPOSE']]
transport = transport[['TRAVEL_ID', 'PAYMENT_NUM', 'PAYMENT_AMT_WON']]
lodging = lodging[['TRAVEL_ID', 'PAYMENT_NUM', 'PAYMENT_AMT_WON']]
activity_consume = activity_consume[['TRAVEL_ID', 'PAYMENT_NUM', 'PAYMENT_AMT_WON']]
activity = activity[['TRAVEL_ID', 'VISIT_AREA_ID', 'ACTIVITY_TYPE_CD']]
visit = visit[['VISIT_AREA_ID', 'VISIT_AREA_NM']]

# âš™ï¸ íƒ€ì… ì •ë¦¬
activity['ACTIVITY_TYPE_CD'] = activity['ACTIVITY_TYPE_CD'].astype(str)

# ğŸ”— ì†Œë¹„ë°ì´í„°ì— ì—¬í–‰ë™ê¸° ë¶™ì´ê¸°
transport = transport.merge(travel, on='TRAVEL_ID', how='left')
lodging = lodging.merge(travel, on='TRAVEL_ID', how='left')
activity_consume = activity_consume.merge(travel, on='TRAVEL_ID', how='left')

# ğŸ”¥ ì—¬í–‰ë™ê¸°ë³„ ì¸ë‹¹ ì†Œë¹„ê¸ˆì•¡ ê³„ì‚°
# ì´ë™
transport['PER_PERSON'] = transport['PAYMENT_AMT_WON'] / transport['PAYMENT_NUM']
transport_grouped = transport.groupby('TRAVEL_PURPOSE')['PER_PERSON'].mean().reset_index()
transport_grouped.rename(columns={'PER_PERSON': 'ì´ë™_1ì¸ë‹¹_í‰ê· _ì†Œë¹„ê¸ˆì•¡'}, inplace=True)

# ìˆ™ë°•
lodging['PER_PERSON'] = lodging['PAYMENT_AMT_WON'] / lodging['PAYMENT_NUM']
lodging_grouped = lodging.groupby('TRAVEL_PURPOSE')['PER_PERSON'].mean().reset_index()
lodging_grouped.rename(columns={'PER_PERSON': 'ìˆ™ë°•_1ì¸ë‹¹_í‰ê· _ì†Œë¹„ê¸ˆì•¡'}, inplace=True)

# í™œë™
activity_consume['PER_PERSON'] = activity_consume['PAYMENT_AMT_WON'] / activity_consume['PAYMENT_NUM']
activity_grouped = activity_consume.groupby('TRAVEL_PURPOSE')['PER_PERSON'].mean().reset_index()
activity_grouped.rename(columns={'PER_PERSON': 'í™œë™_1ì¸ë‹¹_í‰ê· _ì†Œë¹„ê¸ˆì•¡'}, inplace=True)

# ğŸ”— ì´ë™/ìˆ™ë°•/í™œë™ ì†Œë¹„ë¥¼ í•˜ë‚˜ë¡œ í•©ì¹˜ê¸°
consume_summary = transport_grouped.merge(lodging_grouped, on='TRAVEL_PURPOSE')
consume_summary = consume_summary.merge(activity_grouped, on='TRAVEL_PURPOSE')

# ğŸ› ï¸ ì—¬í–‰ë™ê¸° ì½”ë“œ í•œê¸€ ë§¤í•‘
travel_purpose_map = codeB[codeB['cd_a'] == 'MIS'][['cd_b', 'cd_nm']].rename(columns={'cd_b': 'TRAVEL_PURPOSE', 'cd_nm': 'TRAVEL_PURPOSE_NM'})
consume_summary = consume_summary.merge(travel_purpose_map, on='TRAVEL_PURPOSE', how='left')

# ì—¬í–‰ë™ê¸°ê°€ ì—†ëŠ” ë°ì´í„°ëŠ” ì œì™¸
consume_summary = consume_summary.dropna(subset=['TRAVEL_PURPOSE_NM'])
popular_places = popular_places.dropna(subset=['TRAVEL_PURPOSE_NM'])
popular_activities = popular_activities.dropna(subset=['TRAVEL_PURPOSE_NM'])

# ğŸ“ˆ ì†Œë¹„ ì„±í–¥ ê²°ê³¼ ì¶œë ¥
print("\nâœ… [ì—¬í–‰ë™ê¸°ë³„ 1ì¸ë‹¹ í‰ê·  ì†Œë¹„ê¸ˆì•¡]")
print(consume_summary[['TRAVEL_PURPOSE_NM', 'ì´ë™_1ì¸ë‹¹_í‰ê· _ì†Œë¹„ê¸ˆì•¡', 'ìˆ™ë°•_1ì¸ë‹¹_í‰ê· _ì†Œë¹„ê¸ˆì•¡', 'í™œë™_1ì¸ë‹¹_í‰ê· _ì†Œë¹„ê¸ˆì•¡']])

# ğŸ› ï¸ í™œë™ìœ í˜• ë°ì´í„° ë³‘í•©
activity_full = pd.merge(activity, travel, on='TRAVEL_ID', how='left')
activity_full = pd.merge(activity_full, visit, on='VISIT_AREA_ID', how='left')

# í™œë™ìœ í˜• ì½”ë“œ í•œê¸€ ë§¤í•‘
activity_type_map = codeB[codeB['cd_a'] == 'ACT'][['cd_b', 'cd_nm']].rename(columns={'cd_b': 'ACTIVITY_TYPE_CD', 'cd_nm': 'ACTIVITY_TYPE_NM'})
activity_full = activity_full.merge(activity_type_map, on='ACTIVITY_TYPE_CD', how='left')

# ğŸ“Š ì—¬í–‰ë™ê¸°ë³„ ì¸ê¸° ë°©ë¬¸ì§€ëª…, í™œë™ìœ í˜• ë¶„ì„
popular_places = activity_full.groupby(['TRAVEL_PURPOSE', 'VISIT_AREA_NM']).size().reset_index(name='count')
popular_activities = activity_full.groupby(['TRAVEL_PURPOSE', 'ACTIVITY_TYPE_NM']).size().reset_index(name='count')

# ì—¬í–‰ë™ê¸° í•œê¸€ ë¶™ì´ê¸°
popular_places = popular_places.merge(travel_purpose_map, on='TRAVEL_PURPOSE', how='left')
popular_activities = popular_activities.merge(travel_purpose_map, on='TRAVEL_PURPOSE', how='left')

# ê²°ê³¼ ì¶œë ¥
print("\nâœ… [ì—¬í–‰ë™ê¸°ë³„ ì¸ê¸° ë°©ë¬¸ì§€ëª… Top]")
print(popular_places[['TRAVEL_PURPOSE_NM', 'VISIT_AREA_NM', 'count']].sort_values('count', ascending=False).head(20))

print("\nâœ… [ì—¬í–‰ë™ê¸°ë³„ ì¸ê¸° í™œë™ìœ í˜• Top]")
print(popular_activities[['TRAVEL_PURPOSE_NM', 'ACTIVITY_TYPE_NM', 'count']].sort_values('count', ascending=False).head(20))
```
